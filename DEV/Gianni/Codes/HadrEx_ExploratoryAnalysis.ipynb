**Descrição:** Notebook interativo para análise exploratória de dados tabelados. Para um arquivo .csv de input, as análises possíveis são:

- **Exploratory Table:** Atributos estatísticos são calculados para todas as features de acordo com o seu tipo (numérica ou categórica): média, desvio-padrão, mediana, intervalo interquartil, min/max, N. de NaN, Frequências e etc.  

- **Exploratory Plots:** Permite plotar gráficos (histogramas, scatterplots, barplots, crosstables e etc) de forma interativa.  

- **Correlation Plots:** Plota duas matrizes de correlação. A primeira com os coeficientes de Spearman para as features numéricas e a segunda com os valores de Chi-Quadrado para as features categóricas.
